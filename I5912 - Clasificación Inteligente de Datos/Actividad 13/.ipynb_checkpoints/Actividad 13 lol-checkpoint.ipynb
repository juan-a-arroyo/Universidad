##funcion limpiar texto
usando las stopwords revisa cada palabra por una frase que recibe y si la palabra no esta en el stopwords se agrega a una cadena y se regresa como la palabra ya simplificada

```python
def limpiar_texto(desc):
    '''
      Esta función se encarga de limpiar el texto de la columna 'caracteristica1' del DataFrame df.
      Elimina caracteres especiales, números, stop_words y otros caracteres no deseados.
      Args:
          desc (str): La descripción del texto a limpiar.
      Returns:
          str: El texto limpio.
    '''
    correctedText = desc
    # Verificar si la descripción es None o no es una cadena
    if desc is None or not isinstance(desc, str):
        return ""  # Retornar una cadena vacía si desc no es válido
    # Eliminar caracteres especiales y números
    correctedText = re.sub(r'[^a-zA-Z\s]', '', correctedText)
    # Convertir a minúsculas
    correctedText = correctedText.lower()
    # Eliminar stop words
    correctedText = ' '.join([word for word in correctedText.split() if word not in stop_words])
    return correctedText
```

## funcion simplificar texto
me di cuenta tu puedes modificar el numero de palabras para que tenga mejor rendimiento como tu veas pero en general el nombre del objeto esta entre las 2 y 4 primeras palabras asi que cortamos la cadena para optimizar un poco las cosas y sirve que queda mas limpio

```python
def Simplify(texto):
    '''
    Esta función simplifica la cantidad de palabras de un texto
    Args:
        texto (str) texto a modificar la cantidad de palabras
    Returns:
        str: El texto simplificado y limpio.
    '''
    if texto is None or not isinstance(texto, str):
        return ""  # Retornar una cadena vacía si desc no es válido
    # Dividir el texto por espacios y tomar las primeras palabras
    palabras = texto.split()[:5]
    # Unir las dos primeras palabras en un solo string
    simplified = ' '.join(palabras)
    return limpiar_texto(simplified)
```

de esta forma le aplicas la funcion simplify a cada linea del dataframe en su columna caracteristicas y lo agrega a una nueva columna simplified

```python
#de esta forma se crea la nueva columda con nombre simplified con el texto mas limpio
df['simplified'] = df['caracteristicas'].apply(Simplify)
```

## tranformando la informacion a algo que entiendan los algoritmos
yo me base en la columna simplified para clasificar por que es la que tiene los datos mas limpios si encuentras algo mejor avisas o lo pones y explicas por que

```python
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df['simplified']).toarray()
```

#calculando la jerarquia de clusteres para el clustering jerargico
X: Este es el conjunto de datos sobre el cual se realizará el clustering. En tu caso, X es una matriz que representa las características de los textos transformados usando TfidfVectorizer.


```python
Z = linkage(X, method='ward')
```

# conociendo la cantidad de clusteres por cada nivel
con la celda de abajo sacamos las distancias de corte de cada nivel para despues conocer la cantidad de clusteres por cada nivel.

```python
distances = np.linspace(0, Z[-1, 2], num=22)  # Ajusta el número de cortes según lo necesites
clusters_count = []
```

```python
print(distances)
```

ahora por cada distancia en distancia creo una lista con la cantidad de clusters por cada nivel

```python
for d in distances:
    if d > 0:  # Evitar el corte en la distancia 0
        clusters = fcluster(Z, d, criterion='distance')
        clusters_count.append(len(np.unique(clusters)))

print(clusters_count)
```

aqui se ven los valores juntos

```python
# Mostrar el número de clústeres por cada distancia
for d, count in zip(distances[1:], clusters_count):  # Comenzamos desde el primer valor diferente de cero
    print(f'Distancia: {d:.2f}, Número de clústeres: {count}')
```

#clasificando la informacion con un numero de cluster
por cada numero de cluster se ejecuta y se aggrega una nueva columna al dataframe con su cluster

```python
#esto  no es machine learning solo es una forma de clasificar con un algoritmo pero aqui lo deje

for num_clusters in clusters_count:
    print(num_clusters)
    # Clasifica las muestras de texto en clústeres
    cluster_labels = fcluster(Z, num_clusters, criterion='maxclust')

    # Agrega los resultados a un nuevo DataFrame
    df[f'cluster_{num_clusters}'] = cluster_labels
```

## este es machine learning de verdad
el pasado solo era un algoritmo ahora si seria ver bien que palabras dejar y que palabras quitar del stoplist

```python
#este si es el algoritmo que si usa machine learning

for num_clusters in clusters_count:
    print(f'Número de clusters: {num_clusters}')

    # Crear el modelo de AgglomerativeClustering para el número de clusters actual
    agglo = AgglomerativeClustering(n_clusters=num_clusters, linkage='ward')

    # Ajustar el modelo a los datos y obtener las etiquetas de los clusters
    cluster_labels = agglo.fit_predict(X)  # X es tu conjunto de datos

    # Agregar los resultados al DataFrame (df)
    df[f'cluster_{num_clusters}'] = cluster_labels
```
